### Matrix form

```python
obj3: GradedObject = {1: [0, 0], 0: [0, 0]}
diff3 = ChainMap(source=obj3, target=obj3, shift=-1)
diff3.set(1, 0, 0, {(0, 0, 0):  1})
diff3.set(1, 0, 1, {(0, 0, 1):  1})
diff3.set(1, 1, 0, {(0, 0, 2):  1})
diff3.set(1, 1, 1, {(0, 0, 0):  1})
cx3 = ChainComplex(objects=obj3, differential=diff3)
 
print("\n=== Test 3: BETA_PLUS of a 2x2 complex of T0s — expect matrix display ===")
print("\nInput complex:")
print(cx3)
print("\nAfter BETA_PLUS:")
print(BETA_PLUS(cx3))
```
